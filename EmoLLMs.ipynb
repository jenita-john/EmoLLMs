# Introduction:

#### Problem Description:

Existing large language models for sentiment analysis and emotion detection focus solely on classification tasks (e.g., sentiment polarity, emotion categories) while neglecting regression tasks (e.g., sentiment strength, emotion intensity), limiting their ability to capture fine-grained affective features needed for downstream applications.

#### Context of the Problem:

Sentiment analysis and emotion detection is crucial for comprehending people's emotional responses and states towards particular subjects. Affective analysis is fundamental to a wide range of NLP applications including mental health analysis, misinformation detection, and empathetic dialogue systems. As LLMs become increasingly central to NLP pipelines, their inability to perform comprehensive multi-task affective analysis, covering both classification and regression, represents a critical gap, particularly when downstream datasets rely on automated emotion annotation tools.

#### Limitation About other Approaches:

Prior approaches, including PLM-based models (BERT, RoBERTa) and existing LLM fine-tuning methods, focus only individual aspects of affective analysis and classification tasks, failing to generalize across both classification and regression tasks. This is largely caused by the lack of comprehensive affective annotations in most downstream datasets, a gap further compounded by rule-based tools such as VADER and TextBlob, which offer only single-aspect sentiment coverage with limited accuracy.

#### Solution:

The paper proposes a suite of LLMs, instruction-tuning datasets, and an evaluation benchmark for multi-task affective analysis. First, it constructs the multi-task affective analysis instruction dataset (AAID) with 234K samples spanning five classification and regression tasks to support LLM instruction fine-tuning. Based on the AAID dataset, the paper proposes a series of open-sourced instruction-following emotional large language models (EmoLLMs), for comprehensive affective analysis. To evaluate generalizability, the paper introduces the Affective Evaluation Benchmark (AEB), covering 14 affective analysis datasets collected from various platforms and sources.

# Methodology
## **1.   Dataset Construction: Affective Analysis Instruction Dataset (AAID):**

To enable instruction tuning of large language models (LLMs) for affective understanding, the authors construct the Affective Analysis Instruction Dataset, a unified dataset designed for multi-task emotional analysis. AAID is primarily derived from SemEval-2018 Task 1 Affect in Tweets, a benchmark dataset widely used for affective computing.

AAID reformulates traditional affective analysis tasks into an instruction-following format, making it compatible with modern LLM training paradigms. It supports five core tasks:

*   Emotion Intensity Regression (EI-reg) – Predict continuous emotion intensity
* Ordinal Classification of Emotion Intensity (EI-oc) – Classify intensity into ordered categories
* Sentiment Strength Regression (V-reg) – Predict sentiment strength scores
* Sentiment Classification (V-oc) – Classify sentiment polarity (e.g., positive/negative)
* Multi-label Emotion Classification (E-c) – Detect multiple co-occurring emotions

Each data instance is converted into a natural language instruction-response pair, allowing models to learn task generalization through prompt-based learning.


## **2.   EmoLLMs: Multi-task Generative Framework:**

The paper proposes EmoLLMs, a family of instruction-tuned LLMs designed specifically for affective analysis.

Unlike prior approaches that rely on discriminative or regression-based models, EmoLLMs frame affective analysis as a generative problem. Specifically, the model learns a conditional probability distribution: *P*𝜙 (𝑦|𝑥)

where:

* 𝑥: input text (e.g., tweet)
* 𝑦: generated output (emotion label, intensity, or sentiment score)
* 𝜙: pretrained model parameters

This formulation enables a single autoregressive model to handle multiple affective tasks simultaneously. Instead of training separate models per task, EmoLLMs leverage multi-task instruction tuning, allowing knowledge sharing across tasks such as:

* Emotion detection
* Sentiment polarity prediction
* Intensity estimation

This unified framework improves efficiency and generalization while simplifying deployment.

Importantly, EmoLLMs represent one of the first open-source instruction-following LLMs specifically designed for comprehensive affective analysis, enabling downstream applications such as:

* Social media monitoring
* Mental health analysis
* Customer sentiment tracking
* Human-computer interaction systems

## **3. Evaluation Framework: Affective Evaluation Benchmark (AEB):**

To rigorously assess model performance, the authors introduce the Affective Evaluation Benchmark, a large-scale benchmark composed of 14 diverse affective datasets collected from multiple platforms.

The benchmark includes:

* 8 regression tasks
* 6 classification tasks

To evaluate both training effectiveness and generalization, AEB is divided into two subsets:

*AEB-1 (In-domain Evaluation):*

* Contains datasets derived from the same sources as AAID
* Used to measure how well models learn from the training distribution

*AEB-2 (Out-of-domain Evaluation):*
* Contains datasets from different platforms and sources
* Used to evaluate model generalization capability

# Implementation

The EmoLLMs are implemented by fine-tuning several popular open-source LLM architectures using the AAID dataset. The following models are developed:

* LLaMA-based Models
* EmoLLaMA-7B (based on LLaMA 2)
* EmoLLaMA-chat-7B
* EmoLLaMA-chat-13B

The “chat” variants are initialized from models already fine-tuned using Reinforcement Learning from Human Feedback, enabling better instruction-following behavior.

###Other Architectures:
* EmoOPT (based on OPT-13B)
* EmoBLOOM (based on BLOOM-7B)

This diversity ensures robustness across architectures and training strategies.

# Experimental Results

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
tokenizer = AutoTokenizer.from_pretrained('lzw1008/Emot5-large')
model = AutoModelForSeq2SeqLM.from_pretrained('lzw1008/Emot5-large', device_map='auto')
```

```python
def run_task(prompt):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)
    outputs = model.generate(input_ids, max_new_tokens=128)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

## Task 1: Emotion Detection

Detect prescence of emotions in text:

```python
prompt = """Task: Categorize the text's emotional tone as either 'neutral or no emotion' or identify the presence of one or more of the given emotions (anger, anticipation, disgust, fear, joy, love, optimism, pessimism, sadness, surprise, trust).
Text: Whatever you decide to do make sure it makes you happy.
This text contains emotions:"""

print(run_task(prompt))
```

## Task 2: Sentiment Classification

Classify text as positive, negative, or neutral:


```python
prompt = """ Task: Categorize the text into an ordinal class that best characterizes the writer's mental state, considering various degrees of positive and negative sentiment intensity. 3: very positive mental state can be inferred. 2: moderately positive mental state can be inferred. 1: slightly positive mental state can be inferred. 0: neutral or mixed mental state can be inferred. -1: slightly negative mental state can be inferred. -2: moderately negative mental state can be inferred. -3: very negative mental state can be inferred
Text: Beyoncé resentment gets me in my feelings every time. 😩
Intensity Class:"""

print(run_task(prompt))
```

## Task 3: Emotion Intensity (Regression)

Return a score from 0 to 1 for a specific emotion:



```python
prompt = """Task: Assign a numerical value between 0 (least E) and 1 (most E) to represent the intensity of emotion E expressed in the text.
Text: I can't stop smiling today, everything is perfect!
Emotion: joy
Intensity Score:"""

print(run_task(prompt))
```

```python
prompt = """Task: Assign a numerical value between 0 (least E) and 1 (most E) to represent the intensity of emotion E expressed in the text.
Text: Happy Birthday shorty. Stay fine stay breezy stay wavy @daviistuart 😘
Intensity Score:"""

print(run_task(prompt))
```

## Task 4: Batch Test Multiple Sentences



```python
sentences = [
    ("I am so angry I could scream.", "anger"),
    ("This is the best news I've heard all year!", "joy"),
    ("I feel completely empty inside.", "sadness")
]

for text, emotion in sentences:
    prompt = f"""Task: Assign a numerical value between 0 (least E) and 1 (most E) to represent the intensity of emotion E expressed in the text.
Text: {text}
Emotion: {emotion}
Intensity Score:"""
    print(f"Text: {text}")
    print(f"Emotion ({emotion}) intensity: {run_task(prompt)}\n")
```

# Conclusion

This project helped build a clearer understanding of how emotional language models perform across different tasks. By testing EmoLLMs on emotion detection, sentiment classification, and emotion intensity scoring, it became clear that the model can handle clear and strongly expressed emotions quite well.

The results also showed that performance depends a lot on the input. The model works best when the text is short and direct, especially in social-media-style examples. Prompt design also had a strong impact, where clearer and more structured prompts led to more consistent outputs.

When compared with ChatGPT, EmoLLMs appeared more specialized for emotion-related tasks, while ChatGPT was more flexible but sometimes less consistent in scoring intensity or sentiment strength.

However, the model showed limitations when dealing with subtle, mixed, or complex emotions. It also struggled with longer or more formal text, and with culturally nuanced expressions. Overall, the project highlights both the strengths and the practical limitations of emotion-aware language models. <br>


# Future Direction

There are several directions to extend this work. One important step is to test the model on larger and more diverse datasets, including longer texts, different writing styles, and multiple languages, to better evaluate how well it generalizes.

Another improvement area is handling subtle and overlapping emotions, since the model currently struggles with more complex emotional expressions. Fine-tuning on more specific or domain-focused data could help improve consistency.

It would also be useful to compare EmoLLMs with other emotional language models to better understand differences in accuracy, consistency, and performance across tasks.

In addition, future work can focus on real-world applications such as social media monitoring, customer feedback analysis, and mental health-related systems, where emotion understanding is important.

Finally, improving efficiency is important, since these models require significant computational resources. Exploring faster implementations, as well as combining text with other signals like audio or images, could further improve performance and make the models more practical to use.
